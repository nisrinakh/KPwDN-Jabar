```python
import requests
from bs4 import BeautifulSoup
from googlesearch import search
from datetime import datetime

def format_date(date_str):
    try:
        return datetime.strptime(date_str, "%Y-%m-%d").strftime("%d/%m/%Y")
    except ValueError:
        raise ValueError("Tanggal harus dalam format YYYY-MM-DD")

def cari_artikel_berita(kata_kunci, jumlah_hasil=10, tanggal_dari=None, tanggal_sampai=None):
    query = kata_kunci
    print(f"Googling keyword: {query}")  # Debugging
    try:
        # Gunakan stop sebagai pengganti num_results
        search_results = search(query=kata_kunci,num=10, stop=jumlah_hasil, safe="off", lang="id")
        count = 1
        for url in search_results:
            print(f"URL Found {count}: {url}")  # Debugging
            hasil.append(url)
            count = count + 1
        return True
    except Exception as e:
        print(f"Kesalahan dalam pencarian: {e}")
        return False


kata_kunci = input("Masukkan kata atau kalimat yang akan dicari: ")
jumlah_hasil = int(input("Masukkan jumlah hasil yang diinginkan: "))
hasil = []

cari_artikel_berita(kata_kunci, jumlah_hasil)

print("Daftar URL yang mengandung kata-kata tersebut:")
print(hasil)
print(f"")
print(f"URLs Fetching Result:")

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from deep_translator import GoogleTranslator

def get_content_and_title(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    # Mendapatkan judul artikel
    title = soup.find('h1').getText().strip() if soup.find('h1') else 'No title found'

    # Mendapatkan konten artikel
    articles = []
    for paragraph in soup.select('p'):
        article = paragraph.getText().strip()
        articles.append(article)
    contents = " ".join(articles)

    return title, contents

def translate_text(text, translator):
    # Memecah teks menjadi potongan yang lebih kecil jika perlu
    max_length = 5000  # Panjang maksimum potongan teks
    chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]

    translated_chunks = []
    for chunk in chunks:
        try:
            translated_chunk = translator.translate(chunk)
            translated_chunks.append(translated_chunk)
        except Exception:
            # Mengabaikan error tanpa menampilkan pesan
            pass

    # Menggabungkan kembali potongan teks yang diterjemahkan
    translated_text = "".join(translated_chunks)
    return translated_text

def contains_words(string, words):
  return any( word in string for word in words)


# Inisialisasi analyzer VADER dan translator
analyzer = SentimentIntensityAnalyzer()
translator = GoogleTranslator(source='id', target='en')
ignored_urls = ["youtube", "instagram", "facebook", "x.com"]
# Memisahkan daftar URL dan loop untuk setiap URL
url_list = hasil

for url in url_list:
    url = url.strip()  # Menghapus spasi tambahan
    if contains_words(url, ignored_urls):
      continue
    try:
        title, contents = get_content_and_title(url)
    except Exception:
        title = "Error fetching site contents."
        sentiment = "-"
        pass


    if len(contents) < 150:
        sentiment = {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}
        translated_contents = ""
    else:
        try:
            # Menerjemahkan konten dari bahasa Indonesia ke bahasa Inggris
            translated_contents = translate_text(contents, translator)
            # Lakukan analisis sentimen pada teks yang telah diterjemahkan
            sentiment = analyzer.polarity_scores(translated_contents) if translated_contents else {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}
        except Exception:
            print(f"Error fetching site contents.")
            title = "-"
            sentiment = "-"
            pass


    # Print hasil
    print(f"Title: {title}")
    print(f"URL: {url}")
    print(f"Sentiment: {sentiment}")

    # Menampilkan pesan berdasarkan nilai compound
    if sentiment['compound'] > 0:
        print("The news depicts a positive sentiment.")
    elif sentiment['compound'] < 0:
        print("The news depicts a negative sentiment.")
    else:
        print("The news depicts a neutral sentiment.")
    print("|"*1)  # Separator antara hasil untuk setiap URL
```

```python
import requests
from bs4 import BeautifulSoup
from googlesearch import search
from datetime import datetime

def format_date(date_str):
    try:
        return datetime.strptime(date_str, "%Y-%m-%d").strftime("%d/%m/%Y")
    except ValueError:
        raise ValueError("Tanggal harus dalam format YYYY-MM-DD")

def cari_artikel_berita(kata_kunci, jumlah_hasil=10, tanggal_dari=None, tanggal_sampai=None):
    query = kata_kunci
    print(f"Googling keyword: {query}")  # Debugging
    try:
        # Gunakan stop sebagai pengganti num_results
        search_results = search(query=kata_kunci,num=10, stop=jumlah_hasil, safe="off", lang="id")
        count = 1
        for url in search_results:
            print(f"URL Found {count}: {url}")  # Debugging
            hasil.append(url)
            count = count + 1
        return True
    except Exception as e:
        print(f"Kesalahan dalam pencarian: {e}")
        return False


kata_kunci = input("Masukkan kata atau kalimat yang akan dicari: ")
jumlah_hasil = int(input("Masukkan jumlah hasil yang diinginkan: "))
hasil = []

cari_artikel_berita(kata_kunci, jumlah_hasil)

print("Daftar URL yang mengandung kata-kata tersebut:")
print(hasil)
print(f"")
print(f"URLs Fetching Result:")


from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from deep_translator import GoogleTranslator
from google.colab import auth
auth.authenticate_user()
import gspread
from google.auth import default
creds, _ = default()


def get_content_and_title(url):
    # res = requests.get(url)
    # soup = BeautifulSoup(res.text, 'html.parser')

    # # Mendapatkan judul artikel
    # title = soup.find('h1').getText().strip() if soup.find('h1') else 'No title found'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    # Try different methods to find the title
    title = None

    # Method 1: Look for <title> tag in the <head>
    if soup.title:
        title = soup.title.get_text().strip()

    # Method 2: Look for <h1> tag
    if not title and soup.find('h1'):
        title = soup.find('h1').get_text().strip()

    # Method 3: Look for common header tags
    if not title:
        header_tags = ['h1', 'h2', 'h3']
        for tag in header_tags:
            header = soup.find(tag)
            if header:
                title = header.get_text().strip()
                break

    # Method 4: Look for specific classes or IDs (example)
    if not title:
        specific_selectors = [
            '.article-title',  # class example
            '#main-title',     # id example
            '.entry-title'
        ]
        for selector in specific_selectors:
            element = soup.select_one(selector)
            if element:
                title = element.get_text().strip()
                break

    if not title:
      title = 'No title found'

    # Mendapatkan konten artikel
    articles = []
    for paragraph in soup.select('p'):
        article = paragraph.getText().strip()
        articles.append(article)
    contents = " ".join(articles)

    return title, contents

def translate_text(text, translator):
    # Memecah teks menjadi potongan yang lebih kecil jika perlu
    max_length = 5000  # Panjang maksimum potongan teks
    chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]

    translated_chunks = []
    for chunk in chunks:
        try:
            translated_chunk = translator.translate(chunk)
            translated_chunks.append(translated_chunk)
        except Exception:
            # Mengabaikan error tanpa menampilkan pesan
            pass

    # Menggabungkan kembali potongan teks yang diterjemahkan
    translated_text = "".join(translated_chunks)
    return translated_text

def contains_words(string, words):
  return any( word in string for word in words)


# Inisialisasi analyzer VADER dan translator
analyzer = SentimentIntensityAnalyzer()
translator = GoogleTranslator(source='id', target='en')
ignored_urls = ["youtube", "instagram", "facebook", "x.com"]
# Memisahkan daftar URL dan loop untuk setiap URL
url_list = hasil

gc = gspread.authorize(creds)

worksheet = gc.open('Sentiment Analyzer').sheet1

# Pull all 1st column value (kolom no)
first_column_values = worksheet.col_values(1)
row_number = int(first_column_values[-1]) + 1
positive_count = 0
negative_count = 0
neutral_count = 0

for url in url_list:
    url = url.strip()  # Menghapus spasi tambahan
    sentiment_result = ""
    if contains_words(url, ignored_urls):
      continue
    try:
        title, contents = get_content_and_title(url)
        # Split the URL by '//' and then by '/'
        publisher = url.split('//')[1].split('/')[0]

    except Exception:
        title = "Error fetching site contents."
        sentiment = "-"
        # Split the URL by '//' and then by '/'
        publisher = url.split('//')[1].split('/')[0]
        pass
        continue


    if len(contents) < 150:
        sentiment = {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}
        translated_contents = ""
    else:
        try:
            # Menerjemahkan konten dari bahasa Indonesia ke bahasa Inggris
            translated_contents = translate_text(contents, translator)
            # Lakukan analisis sentimen pada teks yang telah diterjemahkan
            sentiment = analyzer.polarity_scores(translated_contents) if translated_contents else {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}
        except Exception:
            print(f"Error fetching site contents.")
            title = "-"
            sentiment = "-"
            pass
            continue
    # getting sentiment result
    if sentiment['compound'] > 0:
        sentiment_result = "Positive"
        positive_count = positive_count + 1
    elif sentiment['compound'] < 0:
        sentiment_result = "Negative"
        negative_count = negative_count + 1
    else:
        sentiment_result = "Neutral"
        neutral_count = neutral_count + 1
    row_result = [row_number, url, publisher, title, sentiment['neg'], sentiment['neu'], sentiment['pos'], sentiment['compound'], sentiment_result]

    # Ignoring trash data
    ignored_keywords = ["ERROR", "error", "403", "Forbidden", "forbidden", "block", "Just a moment", "Cloudflare"]
    if contains_words(row_result[3], ignored_keywords):
      print("Row ignored, title: ", row_result[3])
      continue
    if sentiment['neg'] == 0.0 and sentiment['neu'] == 1.0 and sentiment['pos'] == 0.0 and sentiment['compound'] == 0.0 :
      print("Row ignored, sentiment mostlikely fail to analyze: neg=0, neu=1, pos=0, compound=0")
      continue

    # Append the new row data
    worksheet.append_row(row_result)
    row_number = row_number + 1
    print("New row added successfully:", row_result)

# Get all values in the sheet
all_values = worksheet.get_all_values()

# Get the last row index (1-based index)
last_row_index = len(all_values)

# Get the last row data
last_row_data = all_values[-1]

# Add new data to the additional column
if positive_count > negative_count:
  if positive_count > neutral_count:
    overall_sentiment_result = f"Mostly Positive (Count Pos= {positive_count}; Count Neg= {negative_count}; Count Neu= {neutral_count};)"
  elif neutral_count > positive_count:
    overall_sentiment_result = f"Mostly Neutral (Count Pos= {positive_count}; Count Neg= {negative_count}; Count Neu= {neutral_count};)"
else:
  overall_sentiment_result = f"Mostly Negative (Count Pos= {positive_count}; Count Neg= {negative_count}; Count Neu= {neutral_count};)"
last_row_data.append(overall_sentiment_result)
# Update the last row with the new data
worksheet.update(f'A{last_row_index}:Z{last_row_index}', [last_row_data])
```
